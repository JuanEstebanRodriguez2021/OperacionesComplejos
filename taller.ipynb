```python
import matplotlib.pyplot as plt

def plot_complex_number(z, color='blue', label=None):
    plt.quiver(0, 0, z.real, z.imag, angles='xy', scale_units='xy', scale=1, color=color)
    if label:
        plt.text(z.real, z.imag, f'{z}', fontsize=12, ha='right', color=color)

def plot_operations(z1, z2):
    plt.figure(figsize=(8, 8))
    
    # Dibuja z1 y z2
    plot_complex_number(z1, color='blue', label='z1')
    plot_complex_number(z2, color='red', label='z2')
    
    # Dibuja la suma
    addition = z1 + z2
    plot_complex_number(addition, color='green', label='z1 + z2')
    
    # Dibuja la multiplicaciÃ³n
    multiplication = z1 * z2
    plot_complex_number(multiplication, color='purple', label='z1 * z2')
    
    # Dibuja las lÃ­neas de los ejes
    plt.axhline(0, color='black',linewidth=0.5)
    plt.axvline(0, color='black',linewidth=0.5)
    
    # Configura el grÃ¡fico
    plt.xlabel('Parte Real')
    plt.ylabel('Parte Imaginaria')
    plt.title('NÃºmeros Complejos y Operaciones como Vectores')
    plt.legend()
    plt.grid(True)
    plt.xlim(-10, 10)
    plt.ylim(-10, 10)
    plt.gca().set_aspect('equal', adjustable='box')
    
    plt.show()

# Ejemplo de uso
z1 = 3 + 4j
z2 = 1 + 2j

plot_operations(z1, z2)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def mandelbrot(c, max_iter):
    z = 1
    for n in range(max_iter):
        if abs(z) > 5:
            return n
        z = z*z + c
    return max_iter

# Create a grid of complex numbers
pixels = 800
max_iter = 30
x = np.linspace(-2, 1, pixels)
y = np.linspace(-1.5, 1.5, pixels)
xx, yy = np.meshgrid(x, y)
c = xx + yy*1j

# Compute Mandelbrot set
mandelbrot_set = np.array([mandelbrot(ci, max_iter) for ci in c.ravel()]).reshape(c.shape)

# Plot
plt.imshow(mandelbrot_set, extent=[-2, 1, -1.5, 1.5])
plt.colorbar()
plt.title("Mandelbrot Set")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def julia_set(width, height, x_min, x_max, y_min, y_max, c, max_iter):
    """Generate a Julia set image for a given constant c."""
    
    x = np.linspace(x_min, x_max, width)
    y = np.linspace(y_min, y_max, height)
    X, Y = np.meshgrid(x, y)
    Z = X + 1j * Y
    
    
    img = np.zeros(Z.shape, dtype=float)
    
    
    for i in range(max_iter):
        Z = Z**2 + c
        mask = np.abs(Z) < 10  
        img += mask
        
    # Normalize image
    img = np.log(img + 1) 
    img = img / img.max()  
    
    return img

def plot_julia_set(c, width=800, height=800, x_min=-1.5, x_max=1.5, y_min=-1.5, y_max=1.5, max_iter=100):
    
    img = julia_set(width, height, x_min, x_max, y_min, y_max, c, max_iter)
    plt.imshow(img, extent=(x_min, x_max, y_min, y_max), cmap='inferno')
    plt.colorbar()
    plt.title(f'Julia Set for c = {c}')
    plt.xlabel('Re')
    plt.ylabel('Im')
    plt.show()


c = -0.4 + 0.6j
plot_julia_set(c)


constants = [
    -0.7 + 0.27015j,
    -0.4 + 0.6j,
    0.355 + 0.355j,
    0.285 + 0.01j
]

for c in constants:
    plot_julia_set(c)
```

```python
import numpy as np

def absolute_value_product(z1, z2):
    """Verificar que |z1 * z2| = |z1| * |z2|"""
    product = z1 * z2
    abs_product = abs(product)
    abs_z1 = abs(z1)
    abs_z2 = abs(z2)
    product_of_abs = abs_z1 * abs_z2
    return abs_product, product_of_abs

def conjugate_property(z1, z2):
    """Verificar que el conjugado(z1 + z2) = conjugado(z1) + conjugado(z2)"""
    sum_conjugate = np.conj(z1 + z2)
    conjugate_sum = np.conj(z1) + np.conj(z2)
    return sum_conjugate, conjugate_sum

# Ejemplos de complejos
z1 = 3 + 4j
z2 = 1 - 2j

# Verificar el modulo
abs_product, product_of_abs = absolute_value_product(z1, z2)
print(f"|z1 * z2| = {abs_product}")
print(f"|z1| * |z2| = {product_of_abs}")
print(f"Son iguales? {'Si' if np.isclose(abs_product, product_of_abs) else 'No'}")

# Verificar el conjugado 
sum_conjugate, conjugate_sum = conjugate_property(z1, z2)
print(f"Conjugado of (z1 + z2) = {sum_conjugate}")
print(f"Conjugado de z1 + Conjugado de z2 = {conjugate_sum}")
print(f"Son iguales? {'Si' if np.allclose(sum_conjugate, conjugate_sum) else 'No'}")
```
